#### Importação e visualização do dataset

```python
def baixar_dataset(nome_pasta: str, caminho_kaggle: str):
    """
    Função para baixar um dataset do Kaggle e salvá-lo em uma pasta específica dentro da pasta "datasets".
    
    Parâmetros:
        nome_pasta (str): Nome da pasta onde o dataset será salvo (dentro da pasta "datasets").
        caminho_kaggle (str): Caminho do dataset no Kaggle 
                              (exemplo: "pavansubhasht/ibm-hr-analytics-attrition-dataset").
                              
    Etapas realizadas:
        1. Cria o diretório de destino se não existir.
        2. Baixa o dataset do Kaggle, obtendo o caminho do cache.
        3. Move todos os arquivos do cache para o diretório de destino, "achatando" a estrutura.
        4. Exclui a pasta de cache após mover os arquivos.
    
    Exemplo de uso:
        baixar_dataset_otimizado("ibm-employee-attrition", "pavansubhasht/ibm-hr-analytics-attrition-dataset")
    """
    import os
    import kagglehub  # Biblioteca para download de datasets do Kaggle
    import shutil     # Biblioteca para operações com arquivos e diretórios

    # Define e cria o diretório de destino dentro da pasta "datasets"
    destino = os.path.join("datasets", nome_pasta)
    os.makedirs(destino, exist_ok=True)
    print(f"Diretório criado ou verificado: {destino}")

    # Baixa o dataset do Kaggle e obtém o caminho do cache onde os arquivos foram salvos
    print("Baixando dataset do Kaggle...")
    cache = kagglehub.dataset_download(caminho_kaggle)
    print(f"Dataset baixado no cache: {cache}")

    # Move todos os arquivos do cache para o diretório de destino.
    # Essa abordagem percorre o cache de forma recursiva, mas move apenas os arquivos, sem manter a estrutura de diretórios.
    print(f"Movendo arquivos para o diretório {destino}...")
    for raiz, _, arquivos in os.walk(cache):
        for nome_arquivo in arquivos:
            caminho_origem = os.path.join(raiz, nome_arquivo)
            caminho_destino = os.path.join(destino, nome_arquivo)
            shutil.move(caminho_origem, caminho_destino)
            print(f"Arquivo movido: {caminho_origem} -> {caminho_destino}")

    # Exclui a pasta de cache após mover os arquivos
    shutil.rmtree(cache)
    print(f"Pasta de cache {cache} excluída.")

    print(f"Processo concluído! Dataset disponível em: {destino}")
```

```python
baixar_dataset("ibm-employee-attrition", "pavansubhasht/ibm-hr-analytics-attrition-dataset")
```

#### Contagem de outliers

```python
def contar_outliers(df):
    """
    Conta o número de outliers em cada coluna numérica do DataFrame usando o método Intervalo Interquartil (IQR).
    
    Parâmetros:
    -----------
    df : pandas.DataFrame
        DataFrame contendo os dados a serem analisados
        
    Retorna:
    --------
    dict
        Dicionário com o nome das colunas numéricas como chaves e a quantidade 
        de outliers como valores
    """
    outliers_info = {}
    
    for coluna in df.select_dtypes(include=['number']).columns:
        Q1 = df[coluna].quantile(0.25)
        Q3 = df[coluna].quantile(0.75)
        IQR = Q3 - Q1 # Intervalo interquartil
        
        limite_inferior = Q1 - 1.5 * IQR
        limite_superior = Q3 + 1.5 * IQR
        
        outliers = df[(df[coluna] < limite_inferior) | (df[coluna] > limite_superior)]
        quantidade_outliers = outliers.shape[0]
        
        outliers_info[coluna] = quantidade_outliers
    
    # Ordena o dicionário pela quantidade de outliers (do maior para o menor)
    outliers_ordenados = dict(sorted(outliers_info.items(), key=lambda item: item[1], reverse=True))
    
    return outliers_ordenados

contar_outliers(df)
```
